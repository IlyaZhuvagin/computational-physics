```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
n = 10
x = np.zeros(n)
x[0] = 1
for i in range(1, n):
    x[i] = x[i-1] + (np.cos(x[i-1]) + 1) / np.sin(x[i-1])
y = np.pi
error = (x - y)**2
plt.plot(error)
plt.yscale('log')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
n = 10
x = np.zeros(n)
x[0] = 1
for i in range(1, n):
    x[i] = 0.5 * (x[i-1] + 2 / x[i-1])
y = np.sqrt(2)
error = (x - y)**2
plt.plot(error)
plt.yscale('log')
```
