```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
A = np.array([[0, 1], [0, 0]])
Id = np.eye(2)
plt.figure(figsize=(5, 5))
for x in np.arange(-0.90, 1, 0.01):
    for y in np.arange(-0.90, 1, 0.01):
        z = complex(x, y)
        if np.linalg.norm(np.linalg.inv(z*Id - A), ord = 2) > 100:
            plt.scatter(x, y)          
plt.show()
```
