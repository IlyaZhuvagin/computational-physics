```python
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt
%matplotlib inline
def kron(i, j):
    if i == j:
        return 1
    else:
        return 0
A = np.zeros((32, 32))
for i in range(32):
    for j in range(32):
        A[i, j] = -kron(i,j) + kron(i, j-1) + kron(i, j-2)
print(A)
print(np.linalg.eigvalsh(A))
```

```python
for t in range(51):
    x = np.linalg.norm(expm(A * t), ord = 2)
    plt.scatter(t, x, color = 'MediumSeaGreen')
    print(t, x)
```
